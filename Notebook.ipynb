```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample Data (Replace with actual data from your database or CSV)
students_data = pd.DataFrame({
    'EnrollmentDate': pd.date_range(start='2022-01-01', periods=10, freq='M'),
    'StudentCount': np.random.randint(50, 200, 10)
})

instructors_data = pd.DataFrame({
    'InstructorID': range(1, 21),
    'AnnualSalary': np.random.randint(40000, 120000, 20)
})

courses_data = pd.DataFrame({
    'CourseID': range(1, 11),
    'StudentCount': np.random.randint(5, 50, 10)
})

departments_data = pd.DataFrame({
    'Department': ['Math', 'CS', 'Physics', 'Biology', 'History'],
    'InstructorCount': np.random.randint(5, 15, 5)
})

tuition_data = pd.DataFrame({
    'StudentID': range(1, 21),
    'TotalUnits': np.random.randint(3, 18, 20),
    'TuitionFee': np.random.randint(3000, 12000, 20)
})

# Visualization 1: Student Enrollment Trends
plt.figure(figsize=(10, 5))
plt.bar(students_data['EnrollmentDate'].dt.strftime('%b %Y'), students_data['StudentCount'], color='skyblue')
plt.xlabel('Enrollment Month')
plt.ylabel('Number of Students')
plt.title('Student Enrollment Trends')
plt.xticks(rotation=45)
plt.show()

# Visualization 2: Instructor Salary Distribution
plt.figure(figsize=(10, 5))
plt.hist(instructors_data['AnnualSalary'], bins=10, color='orange', alpha=0.7)
plt.xlabel('Annual Salary ($)')
plt.ylabel('Number of Instructors')
plt.title('Instructor Salary Distribution')
plt.show()

# Visualization 3: Course Enrollment Distribution
plt.figure(figsize=(10, 5))
plt.bar(courses_data['CourseID'], courses_data['StudentCount'], color='green')
plt.xlabel('Course ID')
plt.ylabel('Number of Students')
plt.title('Course Enrollment Distribution')
plt.show()

# Visualization 4: Department-Wise Instructor Count
plt.figure(figsize=(10, 5))
plt.bar(departments_data['Department'], departments_data['InstructorCount'], color='purple')
plt.xlabel('Department')
plt.ylabel('Number of Instructors')
plt.title('Number of Instructors per Department')
plt.show()

# Visualization 5: Tuition vs. Course Units
plt.figure(figsize=(10, 5))
plt.scatter(tuition_data['TotalUnits'], tuition_data['TuitionFee'], color='red')
plt.xlabel('Total Course Units')
plt.ylabel('Tuition Fee ($)')
plt.title('Tuition Fee vs. Course Units')
plt.show()
```
